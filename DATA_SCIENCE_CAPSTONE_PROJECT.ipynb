```python
def standardize_column(df, col_name, new_col_name=None, symbols_to_remove=None):
    if symbols_to_remove is None:
        symbols_to_remove = []
    clean_series = df[col_name].astype(str)
    for symbol in symbols_to_remove:
        clean_series = clean_series.str.replace(symbol, '', regex=False)
    clean_series = clean_series.astype(float)
    mean_val = clean_series.mean()
    std_val = clean_series.std()
    standardized = clean_series.apply(lambda x: (x - mean_val) / std_val)
    if new_col_name is None:
        new_col_name = f"{col_name}_STD"
    df[new_col_name] = standardized
    return df
```

# Data Preprocessing

```python
X = df[['CURR_AGE', 'GENDER', 'ANN_INCOME_INR', 'AGE_CAR_SEG']]
y = df['PURCHASE']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
probs = model.predict_proba(X_test)[:, 1]
results = []
```

```python
cutoff_values = np.linspace(0.2, 0.6, 5)

for cutoff in cutoff_values:
    y_pred = (probs >= cutoff).astype(int)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    results.append({
        'Cutoff': round(cutoff, 2),
        'Accuracy': round(acc, 4),
        'Precision': round(prec, 4),
        'Recall': round(rec, 4),
        'F1_Score': round(f1, 4),
        'Intercept': round(model.intercept_[0], 4),
        'CURR_AGE_COEF': round(model.coef_[0][0], 4),
        'GENDER_COEF': round(model.coef_[0][1], 4),
        'ANN_INCOME_INR_COEF': round(model.coef_[0][2], 8),
        'AGE_CAR_SEG_COEF': round(model.coef_[0][3], 4)
    })
    results_df = pd.DataFrame(results)
print(results_df)
```

```python
import matplotlib.pyplot as plt
```

```python
results_df.plot(x='Cutoff', y=['Accuracy', 'Precision', 'Recall', 'F1_Score'], kind='bar', figsize=(10,6))
plt.title('Metrics vs Cutoff')
plt.ylabel('Score')
plt.xlabel('Cutoff')
plt.xticks(rotation=0)
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
results_df.plot(x='Cutoff', y=['Intercept', 'CURR_AGE_COEF', 'GENDER_COEF', 'ANN_INCOME_INR_COEF', 'AGE_CAR_SEG_COEF'], kind='bar', figsize=(10,6))
plt.title('Cutoff vs Coefficients & Intercept')
plt.ylabel('Coefficient Value')
plt.xlabel('Cutoff')
plt.xticks(rotation=0)
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# Force conversion to numeric (invalid parsing will be set as NaN)
cols_to_convert = ['CURR_AGE', 'GENDER', 'ANN_INCOME_INR', 'AGE_CAR_SEG']
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Now you can safely apply the coefficients
intercept = -1.6051
CURR_AGE_COEF = -0.0099
GENDER_COEF = 0.2392
ANN_INCOME_INR_COEF = 0.423691
AGE_CAR_SEG_COEF = 0.951

df['Z'] = (
    intercept +
    (CURR_AGE_COEF * df['CURR_AGE']) +
    (GENDER_COEF * df['GENDER']) +
    (ANN_INCOME_INR_COEF * df['ANN_INCOME_INR']) +
    (AGE_CAR_SEG_COEF * df['AGE_CAR_SEG'])
)

# Logistic Regression Probability
df['Probability'] = 1 / (1 + np.exp(-df['Z']))
```
